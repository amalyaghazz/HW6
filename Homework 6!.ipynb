```r
#Question 2
#a
id <- c(1,2,3,4,5)
Name <- c("Peter", "Amy", "Ryan", "Gary", "Michelle")
Salary <- c(623.30, 515.20, 611.00, 729.00, 843.25)

df <- data.frame(id, Name, Salary)

print (df)

#add a column
#b
df$department<-c("IT","Finance","Computer Science","Marketing","Data Analytics")

print (df)
```

```r
#extracting rows 1,3,5 from columns 2,3
#c
df[c(1,3,5),c(2,3)]
```

```r
#barplot
#d

x<-df[c(1,4,5),"salary"]
x
barplot(x,
    names.arg = c("Peter","Gary","Michelle"),
    main = "Salary",
    xlab = "name",
    ylab = 'salary',
    col="darkred",
    )
```

```r
#pie chart
#e
a=max(salary)
b=min(salary)
c=median(salary)
salary1<-c(a,b,c)
percentage<-round(100*c(a,b,c)/sum(c(a,b,c)))
pie(salary1,main="salary", labels=paste(c("max","min","median"),sep=" ",percentage,"%"))
```
